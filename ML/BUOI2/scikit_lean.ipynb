# Build binary features by creating thresholds
### Feature: thuộc tính 
Các thông số mô tả một đối tượng nào đó. Ví dụ thuộc tính màu da, chiều cao, cân nặng để mô tả một đối tượng con người
### Threshold: ngưỡng
Giá trị xác định các mức độ các nhau. Ví dụ 4 điểm là ngưỡng đậu hoặc rớt một môn
### Binarize
Quá trình chia dữ liệu dạng chữ thành 2 label: 0 và 1


```python
# creat a new column named "Result"
lb = preprocessing.LabelBinarizer()
data["Result"] = lb.fit_transform(data["Opportunity Result"])

data.head()
```

# with threshold

```python
data.describe()
```

```python
# dùng mức 75$ để chia làm 2 lable
# Elapsed Days In Sales Stage và Ratio Days Identified To Total Days

binarizer_elapsedDays = Binarizer(threshold = 65) # binarizer_elapsedDays: Bạn đang tạo một đối tượng Binarizer với ngưỡng là 65 cho cột "Elapsed Days In Sales Stage". Điều này có nghĩa là tất cả các giá trị lớn hơn 65 sẽ được chuyển thành 1, và tất cả các giá trị nhỏ hơn hoặc bằng 65 sẽ được chuyển thành 0.
binarizer_ratioDaysIden = Binarizer(threshold = 0.19) # binarizer_ratioDaysIden: Bạn cũng tạo một đối tượng Binarizer khác với ngưỡng là 0.19 cho cột "Ratio Days Identified To Total Days". Điều này có nghĩa là tất cả các giá trị lớn hơn 0.19 sẽ được chuyển thành 1, và tất cả các giá trị nhỏ hơn hoặc bằng 0.19 sẽ được chuyển thành 0.
```

```python
elapsedDays = data ["Elapsed Days In Sales Stage"].values.reshape(1,-1)
ratioDaysIden = data["Ratio Days Identified To Total Days"].values.reshape(1,-1)

#reshape (1,-1) chuyển cột giá trị thành vector giá trị
```

```python
elapsedDays
#Đây là một mảng 2D có một hàng và nhiều cột. Mỗi phần tử trong hàng này là một giá trị của biến "Elapsed Days In Sales Stage" cho từng mẫu dữ liệu.
```

```python
elapsedDays_binary = binarizer_elapsedDays.fit_transform(elapsedDays)
ratioDaysIden_binary = binarizer_ratioDaysIden.fit_transform(ratioDaysIden)
```

```python
elapsedDays_binary
```

# Build Linear Model with Outliers
- Outlier của một feature là giá trị khác xa hoàn toàn với các giá trị còn lại của feature đó
- Outlier ảnh hưởng đến độ chính xác của mô hình máy học

```python
data.head()
```

```python
## similarly , ve encode the rest of the categorical features
le = preprocessing.LabelEncoder() # Đoạn mã này sử dụng LabelEncoder từ module preprocessing của thư viện sklearn để mã hóa các đặc trưng phân loại (categorical features).
                                  #  Tạo một đối tượng LabelEncoder để mã hóa các giá trị phân loại thành các số nguyên.
```

```python
#bạn có thể sử dụng phương thức fit_transform() để áp dụng mã hóa cho các cột chứa các đặc trưng phân loại. Phương thức này sẽ học các nhãn duy nhất trong mỗi cột và gán cho mỗi nhãn một số nguyên duy nhất.

data['Supplies_Subgroup_Encoded'] = le.fit_transform(data['Supplies Subgroup'])
data['Supplies_Group_Encoded'] = le.fit_transform(data['Supplies Group'])
data['Region_Encoded'] = le.fit_transform(data['Region'])
data['Route_To_Market_Encoded'] = le.fit_transform(data['Route To Market'])
data['Competitor_Type_Encoded'] = le.fit_transform(data['Competitor Type'])
```

```python
data.head()
```

```python
final_data = data.drop(['Supplies Subgroup', 'Supplies Group', 'Region', 'Route To Market', 'Competitor Type', 'Opportunity Result'], axis=1)
```

```python
final_data.head()
```

```python
# Tạo biến y chứa cột "result" là nhãn của dữ liệu
y = final_data["Result"]
# Tạo biến x chứa tất cả các cột trừ cột "result", là các đặc trưng của dữ liệu
X = final_data.loc[:,final_data.columns != "Result"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print (X_train. shape, y_train. shape)
print (X_test.shape, y_test.shape)
```

```python
# build the model vithout removing any outliers
logreg = LogisticRegression ()
logreg.fit(X_train, y_train)
```

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
# value of 0 means missing values for the following columns
# Plasma glucose concentration (Glucose)
# Diastotic blood pressure (BloodPressure)
# Triceps skinfold thickness (SkinThickNess)
# 2-hour serum insulin (Insulin)
# Body mass index (BMI)

# Giá trị bằng 0 có nghĩa là thiếu giá trị cho các cột sau
# Nồng độ glucose huyết tương (Glucose)
# Huyết áp tâm trương (Huyết áp)
# Độ dày nếp da cơ tam đầu (SkinThickNess)
# Insulin huyết thanh 2 giờ (Insulin)
#Chỉ số khối cơ thể (BMI)

"""Đoạn comment mà bạn cung cấp giải thích rằng giá trị bằng 0 trong các cột tương ứng có ý nghĩa là thiếu dữ liệu cho các thuộc tính sau:

Nồng độ glucose huyết tương (Glucose): Giá trị 0 có thể đại diện cho việc không có dữ liệu được thu thập hoặc dữ liệu bị thiếu cho thuộc tính này.
Huyết áp tâm trương (Huyết áp): Tương tự, giá trị 0 có thể chỉ ra rằng không có thông tin về huyết áp được thu thập hoặc dữ liệu bị thiếu.
Độ dày nếp da cơ tam đầu (SkinThickNess): Giá trị 0 có thể đại diện cho việc không có dữ liệu về độ dày nếp da được thu thập hoặc dữ liệu bị thiếu.
Insulin huyết thanh 2 giờ (Insulin): Tương tự như các thuộc tính trước, giá trị 0 có thể chỉ ra rằng không có thông tin về insulin được thu thập hoặc dữ liệu bị thiếu.
Chỉ số khối cơ thể (BMI): Một lần nữa, giá trị 0 có thể đại diện cho việc không có thông tin về chỉ số khối cơ thể hoặc dữ liệu bị thiếu."""
```

```python
# mark zreo values as NaN (Not a Number)  0 trong các cột "Glucose", "BloodPressure", "SkinThickness", "Insulin", và "BMI" của DataFrame data1 thành NaN (Not a Number), để chỉ ra rằng các giá trị đó là dữ liệu bị thiếu hoặc không có ý nghĩa.

data1[["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]] = \
    data1[["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"] ].replace (0, np.NaN) # .replace(0, np.NaN): Sử dụng phương thức replace để thay thế mọi giá trị 0 trong các cột đã chọn thành NaN (Not a Number). Điều này giúp đánh dấu các giá trị 0 là dữ liệu bị thiếu hoặc không có ý nghĩa.
```
