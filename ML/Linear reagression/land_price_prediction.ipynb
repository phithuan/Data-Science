```python
# Import thư viện
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dữ liệu
diện_tích = np.array([6, 8, 10, 14, 19]).reshape(-1, 1)#reshape được sử dụng để thay đổi hình dạng của mảng. -1 cho biết số hàng sẽ được xác định tự động dựa trên độ dài của mảng và số cột. 1 chỉ ra rằng mỗi hàng sẽ chỉ có một cột, do đó biến mảng một chiều thành mảng hai chiều (cụ thể là cột vector).
giá_đất = np.array([72, 90, 130, 160, 170])

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(diện_tích, giá_đất)

# Biểu diễn dữ liệu lên đồ thị
plt.scatter(diện_tích, giá_đất, color='blue', label='Dữ liệu')
plt.plot(diện_tích, model.predict(diện_tích), color='red', label='Đường hồi quy')
plt.xlabel('Diện tích (m²)')
plt.ylabel('Giá đất ($)')
plt.title('Biểu đồ hồi quy tuyến tính')
plt.legend()
plt.show()
```

```python
# Vẽ đường hồi quy tuyến tính 
print(f"Phương trình hồi quy: Giá đất = {model.coef_[0]:.2f} * Diện tích + {model.intercept_:.2f}")
```

```python
# Dự đoán giá đất khi diện tích là 10m² và 21m²
diện_tích_dự_đoán = np.array([10, 21]).reshape(-1, 1)
giá_đất_dự_đoán = model.predict(diện_tích_dự_đoán)
print(f"Giá đất dự đoán (10m²): ${giá_đất_dự_đoán[0]:.2f}")
print(f"Giá đất dự đoán (21m²): ${giá_đất_dự_đoán[1]:.2f}")
```
