```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = [[6], [8], [10], [14], [18]]
y = [[7], [9], [13], [17.5], [18]]

plt.figure(figsize=(8, 5))
plt.xlabel('Đường kính (inches)')
plt.ylabel('Giá bán tương ứng')
plt.axis([0, 25, 0, 30])
plt.grid(True) # Thêm lưới vào đồ thị để dễ quan sát hơn.

X2 = [[0], [10], [14], [25]]
model = LinearRegression() #  Khởi tạo mô hình hồi quy tuyến tính.
model.fit(X, y) # Huấn luyện mô hình với dữ liệu X và y
y2 = model.predict(X2) # Dự đoán giá bán tương ứng cho X2 sử dụng mô hình đã huấn luyện.
""", nếu bạn đã huấn luyện một mô hình hồi quy tuyến tính (LinearRegression) với dữ liệu đào tạo, 
bạn có thể sử dụng hàm predict để dự đoán giá trị đầu ra cho một tập dữ liệu mới chưa từng thấy. 
Điều này giúp bạn hiểu được mô hình dự đoán như thế nào dựa trên dữ liệu mà nó đã học."""

"""Vẽ dữ liệu và đường hồi quy lên đồ thị. 
'k.' chỉ định vẽ các điểm dữ liệu dưới dạng chấm màu đen và 
'g-' chỉ định vẽ đường hồi quy màu xanh lá."""
plt.plot(X, y, 'k.')    
plt.plot(X2, y2, 'g-')
plt.show()
```

```python
# Giả sử đây là mô hình đã được huấn luyện với dữ liệu X và y
model = LinearRegression()
model.fit(X, y)  # X và y là dữ liệu đã được sử dụng để huấn luyện mô hình

def predict_price(diameter):
    """
    Dự đoán giá bán dựa trên đường kính nhập vào.
    
    Parameters:
    diameter (float): Đường kính của sản phẩm.
    
    Returns:
    float: Giá bán được dự đoán.
    """
    return model.predict([[diameter]])[0][0]

# Ví dụ nhập vào đường kính
diameter_input = float(input("Nhập vào đường kính của chiếc bánh: "))
predicted_price = predict_price(diameter_input)
print(f"Giá bán dự đoán cho đường kính của bánh {diameter_input} inches là: ${predicted_price:.2f}")
```
