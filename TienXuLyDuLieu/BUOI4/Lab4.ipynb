###### Image Resizing: Đảm bảo tất cả hình ảnh có kích thước nhất định để dễ dàng xử lý. OpenCV hoặc thư viện PIL (Pillow) có thể được sử dụng để thực hiện thay đổi kích thước hình ảnh.
###### Noise Reduction: Áp dụng các kỹ thuật lọc, như lọc Gauss hoặc Median, để giảm thiểu nhiễu trong hình ảnh.
###### Color Correction: Cân chỉnh màu sắc của hình ảnh để đảm bảo màu sắc chuẩn xác.
###### Segmentation: Phân đoạn hình ảnh để phân biệt và xác định vùng quan tâm, ví dụ như vùng chứa trái cây và nền.
###### Visualization: Sử dụng thư viện như Matplotlib hoặc Seaborn để hiển thị hình ảnh và thống kê các phân phối màu sắc, kích thước, hoặc bất kỳ thông tin hữu ích nào khác.
###### Data Split: Phân chia tập dữ liệu thành tập huấn luyện và tập kiểm tra để huấn luyện và kiểm định mô hình.
###### Augmentation: Tăng cường dữ liệu để mở rộng tập dữ liệu huấn luyện. Các kỹ thuật như xoay, co giãn, cắt, hoặc phản chiếu có thể được áp dụng.
###### Normalization: Chuẩn hóa hình ảnh để đưa các giá trị pixel về khoảng giá trị cụ thể, thích hợp cho việc huấn luyện mô hình.
###### Feature Extraction: Trích xuất đặc trưng từ hình ảnh, như sử dụng mạng nơ-ron tích chập (CNN) hoặc các phương pháp khác để xác định và rút trích các đặc trưng đặc biệt.

##### Trong hàm segment_image(image), chúng ta sử dụng phương pháp ngưỡng đơn giản để tách ảnh thành các phần đối tượng (foreground) và phần nền (background). Cụ thể, hàm này sử dụng phương pháp ngưỡng để chuyển đổi ảnh màu sang ảnh xám và thực hiện ngưỡng tại mức 127. Kết quả sau khi áp dụng phương pháp ngưỡng sẽ là một ảnh nhị phân với các pixel đen hoặc trắng, giúp tách biệt phần vật thể với nền. Trong trường hợp sử dụng cv2.THRESH_OTSU, ngưỡng được tự động chọn dựa trên phân phối pixel của ảnh, giúp tăng độ chính xác của việc phân đoạn ảnh.

```python
def segment_image(image):
    # Apply image segmentation techniques as needed
    # For example, you can use cv2.threshold() for simple thresholding
    """image: Đây là ảnh đầu vào cần được phân đoạn.
127: Đây là giá trị ngưỡng. Bất kỳ pixel nào có giá trị lớn hơn ngưỡng sẽ được coi là phần vật thể.
255: Giá trị tối đa mà các pixel được gán nếu chúng vượt qua ngưỡng.
cv2.THRESH_BINARY | cv2.THRESH_OTSU: Là cờ hoặc (|) giữa hai cấp độ ngưỡng. Trong đó:
cv2.THRESH_BINARY: Ngưỡng nhị phân cơ bản.
cv2.THRESH_OTSU: Đây là cách chọn ngưỡng tự động dựa trên phân phối pixel. Nó sẽ xác định ngưỡng tối ưu dựa trên histogram của ảnh."""
    ret, segmented = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY|cv2.THRESH_OTSU) 
    return segmented

otsu_thresh = segment_image(img_reduce_noise)

cv2.imshow("Reduced Noise Image", img_reduce_noise) # hình củ đã giảm noise(xám)
cv2.imshow('Otsu Thresholded Image', otsu_thresh)  # Hiển thị ảnh đã otsu_thresh
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import os

def apply_otsu_threshold_and_save(image_path, output_folder):
    # Đảm bảo tồn tại thư mục đầu ra
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Đọc ảnh và chuyển sang ảnh xám
    image = cv2.imread(image_path, 0)
    image = cv2.resize(image, (0, 0), fx=4, fy=4)

    # Áp dụng ngưỡng Otsu
    ret, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Lưu ảnh vào thư mục đầu ra
    output_path = os.path.join(output_folder, f"otsu_{os.path.basename(image_path)}")
    """output_path: Đây là biến lưu trữ đường dẫn đầy đủ của tệp hình ảnh đầu ra.
    -os.path.join(output_folder, f"otsu_{os.path.basename(image_path)}"): Đây là cách để tạo đường dẫn đầy đủ của tệp hình ảnh đầu ra.
    -output_folder: Đây là đường dẫn tới thư mục chứa các tệp hình ảnh đầu ra.
    -f"otsu_{os.path.basename(image_path)}": Tạo tên tệp mới cho hình ảnh đã xử lý.
    -os.path.basename(image_path): Lấy tên tệp hình ảnh từ đường dẫn image_path.
    -f"otsu_{os.path.basename(image_path)}": Đặt tên cho tệp đầu ra bằng cách thêm tiền tố "otsu_" vào tên gốc của tệp hình ảnh."""

    cv2.imwrite(output_path, otsu_thresh)

    return output_path




"""
-input_folder là đường dẫn đến thư mục chứa ảnh.
-output_folder là đường dẫn đến thư mục đầu ra để lưu ảnh sau khi áp dụng ngưỡng Otsu.
-image_files lưu trữ danh sách các tệp ảnh trong thư mục đầu vào, chỉ chọn các tệp có định dạng hình ảnh (.jpg, .jpeg, .png).
-Đoạn code trong vòng lặp for duyệt qua từng tệp ảnh trong image_files:
Sử dụng hàm apply_otsu_threshold_and_save() để áp dụng ngưỡng Otsu vào từng tệp ảnh.
Lưu ảnh sau khi áp dụng ngưỡng Otsu vào thư mục đầu ra output_folder.
In ra đường dẫn của ảnh đã được xử lý."""
# Thư mục chứa ảnh và thư mục đầu ra
input_folder = "C:\\Users\\Admin\\Desktop\\GKD\\Data-science\\TienXuLyDuLieu\\BUOI4\\Fruits360_Lab4\\Training\\Apple Braeburn"
output_folder = "C:\\Users\\Admin\\Desktop\\GKD\\Data-science\\TienXuLyDuLieu\\BUOI4\\output_image_Apple"

# Lấy danh sách các tập tin ảnh trong thư mục đầu vào
image_files = [os.path.join(input_folder, file) for file in os.listdir(input_folder) if file.endswith(('.jpg', '.jpeg', '.png'))]
""" os.listdir(input_folder): Sử dụng hàm os.listdir() để lấy danh sách tất cả các tệp trong thư mục được chỉ định bởi input_folder.

    for file in os.listdir(input_folder): Duyệt qua từng tệp trong danh sách các tệp được lấy từ os.listdir().

    if file.endswith(('.jpg', '.jpeg', '.png')): Kiểm tra xem tên tệp có kết thúc bằng một trong các phần mở rộng ('.jpg', '.jpeg', '.png') hay không.

    os.path.join(input_folder, file): Sử dụng os.path.join() để nối input_folder với tên tệp [input_folder with file name] x (được lấy từ os.listdir()) để tạo đường dẫn đầy đủ đến từng tệp hình ảnh thỏa mãn điều kiện."""

# Áp dụng ngưỡng Otsu và lưu ảnh vào thư mục đầu ra
for image_file in image_files:
    output_path = apply_otsu_threshold_and_save(image_file, output_folder)
    print(f"Saved Otsu thresholded image: {output_path}")
```
