```python
import pandas as pd
```

```python
df_example =pd.DataFrame(
    data={
        "height":[147, 150, 153, 158, 163, 165, 168, 178, 173, 175, 178, 180, 183],
        "weight":[49, 50, 51, 54, 58, 59, 60, 62, 63, 64, 66, 67, 68],
        "height_2":[110, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183],
        "weight_2":[49, 98, 51, 54, 58, 59, 60, 62, 63, 64, 66, 67, 68],
    }
)
df_example
```

*  TH1 (trái): dữ liệu trong cột height làm đầu vào, trong cột weight làm nhãn.
*  TH2 (giữa): dữ liệu trong cột height_2 làm đầu vào, trong cột weight làm nhãn.
*  TH3 (phải): dữ liệu trong cột height làm đầu vào, trong cột weight_2 làm nhãn

```python
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def ftl_linear_regression_and_visualize(df, input_col, label_col):
    # Tạo một đối tượng LinearRegression (Hồi quy tuyến tính).
    lin_reg = LinearRegression(fit_intercept=True)
    
    # Sử dụng phương thức fit để phù hợp mô hình với dữ liệu.
    lin_reg.fit(df[[input_col]], df[label_col])
    
    # Trích xuất hệ số w1 (slope) và w0 (intercept) từ mô hình.
    w1 = lin_reg.coef_# Hệ số này thể hiện mức độ biến đổi của biến phụ thuộc (label_col) khi biến đầu vào (input_col) tăng lên một đơn vị. Nếu w1 dương, tức là có một mối tương quan tích cực giữa biến đầu vào và biến đầu ra, tức là khi biến đầu vào tăng, biến đầu ra cũng tăng.
    w0 = lin_reg.intercept_#Hệ số này thể hiện giá trị của biến phụ thuộc (label_col) khi biến đầu vào (input_col) bằng 0. Nó thường là một điểm trên trục y khi bạn vẽ đường hồi quy tuyến tính.
    
    # Vẽ biểu đồ #visualize
    plt.plot(df[input_col], df[label_col], "ro", label="data")
    plt.axis([105, 290, 45, 75])  # Đặt giới hạn trục x và trục y trên biểu đồ.
    plt.xlabel("Height (cm)")     # Nhãn trục x.
    plt.ylabel("Weight (kg)")     # Nhãn trục y.
    plt.ylim(45, 90)              # Đặt giới hạn trục y.
    
    # Vẽ đường hồi quy tuyến tính.
    plt.plot([105, 190], [w1*105 + w0, w1*190 + w0], label="fitted line")
    # [105, 190]: Đây là danh sách các giá trị trên trục x của hai điểm mà bạn muốn nối để vẽ đoạn đường
    # [w1*105 + w0, w1*190 + w0] Đây là danh sách các giá trị trên trục y tương ứng với hai điểm trên trục x đã nêu ở trên. 
    # Điều này được tính toán bằng cách sử dụng hệ số w1 (slope) và w0 (intercept) đã trích xuất từ mô hình hồi quy tuyến tính. Cụ thể, bạn tính giá trị của biến phụ thuộc (label_col) tại hai điểm x được chọn.
    plt.legend()  # Hiển thị chú thích (legend) trên biểu đồ.
    
    plt.figure(figsize=(17, 6))  # Đặt kích thước của biểu đồ.

# Tạo subplot 1 trong 3 cột của một biểu đồ lớn.
plt.subplot(1, 3, 1)
ftl_linear_regression_and_visualize(df_example, input_col="height", label_col="weight")

# Tạo subplot 2 trong 3 cột của một biểu đồ lớn.
plt.subplot(1, 3, 2)
ftl_linear_regression_and_visualize(df_example, input_col="height_2", label_col="weight")

# Tạo subplot 3 trong 3 cột của một biểu đồ lớn.
plt.subplot(1, 3, 3)
ftl_linear_regression_and_visualize(df_example, input_col="height", label_col="weight_2")
    
```
