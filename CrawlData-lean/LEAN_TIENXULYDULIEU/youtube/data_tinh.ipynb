```python
import requests
from bs4 import BeautifulSoup
#ctrl F để tìm kím streen web
```

```python
url = requests.get('https://hocpython.org/data/')
soup = BeautifulSoup(url.text, "html.parser")
```

```python
# Tìm tất cả các hàng (rows) trong bảng có class là "stock_online"
col = soup.find_all("div", class_='col-inner')
print(col)
```

```python
test = soup.find_all(id="example")
print(test[0])
```

```python
for i in test:
  cells = i.find_all("tr")
  print(cells)
```

```python
# Lặp qua danh sách dữ liệu và hiển thị thông tin
for row in cells:
    cells1 = row.find_all('td')
    
    # Trích xuất dữ liệu từ các ô
    Tên = cells1[0].get_text()
    Vịtrí = cells1[1].get_text()
    Địađiểm = cells1[2].get_text()
    Tuổi = cells1[3].get_text()
    Ngàybắtđầu = cells1[4].get_text()
    
    # Hiển thị thông tin
    print(f"Tên: {Tên}")
    print(f"Vị trí: {Vịtrí}")
    print(f"Địa điểm: {Địađiểm}")
    print(f"Tuổi: {Tuổi}")
    print(f"Ngày bắt đầu: {Ngàybắtđầu}")
    print("\n")
```
