```python
#import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\\GKD\\Data-Science\\CrawlData-lean\\test\\DanhSachCoPhieuTrucTuyenclear.csv")
df.head(30)
```

```python
df.describe()

#count: số lượng các quan sát (số lượng dòng).
#mean: giá trị trung bình của các quan sát.
#std: độ lệch chuẩn của các quan sát.
#min: giá trị nhỏ nhất của các quan sát.
#25%: phần vị 25% của dữ liệu.
#50%: phần vị 50% của dữ liệu (tương đương với giá trị trung bình).
#75%: phần vị 75% của dữ liệu.
#max: giá trị lớn nhất của các quan sát.
```

```python
# Trực quan hóa
plt.figure(figsize=(16, 8))  # Đặt kích thước hình là 16x8

# Biểu đồ giá
plt.subplot(2, 2, 1)
sns.barplot(x='Mã CK', y='Giá', data=df)
plt.title('Biểu đồ giá')

# Biểu đồ khối lượng
plt.subplot(2, 2, 2)
sns.barplot(x='Mã CK', y='Khối lượng', data=df)
plt.title('Biểu đồ khối lượng')

# Biểu đồ tham chiếu
plt.subplot(2, 2, 3)
sns.barplot(x='Mã CK', y='Tham chiếu', data=df)
plt.title('Biểu đồ tham chiếu')

# Biểu đồ +/- (percent change)
plt.subplot(2, 2, 4)
sns.barplot(x='Mã CK', y='+/-', data=df)
plt.title('Biểu đồ +/-: (Tăng/Giảm giá): là mức thay đổi giá sao với Giá tham chiếu.')

plt.tight_layout()
plt.show()
```
