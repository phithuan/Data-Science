```python
train_df["Survived"] = train_df["Survived"].astype("category")
# astype("category") là kiểu dử liệu phân loại     
"""Dữ liệu phân loại là dữ liệu bao gồm các danh mục hoặc nhãn. Loại dữ liệu này được sử dụng để biểu thị dữ liệu không phải là số hoặc dữ liệu có số lượng giá trị có thể hạn chế.

Chuyển đổi một cột thành phân loại có thể hữu ích trong trường hợp bạn muốn nhóm dữ liệu hoặc giảm mức sử dụng bộ nhớ. Khi bạn chuyển đổi một cột thành kiểu dữ liệu phân loại, gấu trúc sẽ tạo ánh xạ giữa các giá trị danh mục và mã số nguyên của chúng, có thể được sử dụng để biểu thị dữ liệu hiệu quả hơn.

Ví dụ: nếu bạn có một cột gồm các chuỗi đại diện cho tên của các quốc gia, bạn có thể chuyển đổi cột đó thành kiểu dữ liệu phân loại để tiết kiệm bộ nhớ và giúp nhóm dữ liệu theo quốc gia dễ dàng hơn."""
```
